```python
def jaccard(a, b):
    a = set(a)
    b = set(b)
    return len(a.intersection(b)) / len(a.union(b))

def intergroup(g, sim_func):
    n = len(g)
    sim = 0
    for i in range(n):
        for j in range(i+1, n):
            sim += sim_func(g[i], g[j])
    sim = sim / (n * (n-1) / 2)
    return round(sim, 2)

def intragroup(g1, g2, sim_func):
    sim = 0
    for i in range(len(g1)):
        for j in range(len(g2)):
            sim += sim_func(g1[i], g2[j])
    sim = sim / (len(g1) * len(g2))
    return round(sim, 2)

def intergroup_sims(g, sim_func):
    n = len(g)
    sims = []
    for i in range(n):
        for j in range(i+1, n):
            sims.append(sim_func(g[i], g[j]))
    return sims

def intragroup_sims(g1, g2, sim_func):
    sims = []
    for i in range(len(g1)):
        for j in range(len(g2)):
            sims.append(sim_func(g1[i], g2[j]))
    return sims
```
